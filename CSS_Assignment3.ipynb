2. Explain the difference between Absolute and Relative positioning.

    - position: relative;

- The element is positioned relative to its normal position in the document flow.

- It still occupies space in the layout.

- You can use top, right, bottom, and left to nudge it from where it would normally be

- It will be in the documentflow and affects the layout

.relative-box {
  position: relative;
  top: 20px;
  left: 10px;
}
This moves the element 20px down and 10px right from where it would normally appear, but its original space is still preserved.

     - position: absolute;

- The element is positioned relative to the nearest positioned ancestor (i.e., the closest parent that has position: relative, absolute, or fixed).

- If no such ancestor exists, it’s positioned relative to the <html> (the page itself).

- It is removed from the normal document flow—other elements act like it’s not there.

- It will not affect the layout

.absolute-box {
  position: absolute;
  top: 20px;
  left: 10px;
}
This places the element 20px from the top and 10px from the left of its positioned ancestor (or the page if there is none).

5. Explain the z-index, with a code example.

    -  
      - It's used when elements overlap.
      - z-index determines the stacking order of elements along the z-axis.
      - Higher values place elements on top of lower ones.
      - Lower values place elements below of upper ones.
      - Only works on elements with a position value other than static (i.e.,
       relative, absolute, fixed, or sticky).
